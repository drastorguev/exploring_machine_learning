```python
from nltk.corpus import stopwords # Import the stop word list
print stopwords.words("english") 

def review_to_words( raw_review ):       
    letters_only = re.sub("[^a-zA-Z]", " ", raw_review) 
    words = letters_only.lower().split()                             
    stops = set(stopwords.words("english"))                  
    meaningful_words = [w for w in words if not w in stops]   
    return( " ".join( meaningful_words ))   
```

```python
import gensim
import re
from nltk.tokenize import word_tokenize
# import nltk
# nltk.download('punkt')
gen_docs = [[w.lower() for w in word_tokenize(review_to_words(text))] for text in newsgroups_train.data[:10000]]
# print gen_docs
```

```python
dictionary = gensim.corpora.Dictionary(gen_docs)
print dictionary[5]
print("Number of words in dictionary:",len(dictionary))
# for i in range(len(dictionary)):
#     print(i, dictionary[i])
```

```python
corpus = [dictionary.doc2bow(gen_doc) for gen_doc in gen_docs]
# print corpus
```

```python
tf_idf = gensim.models.TfidfModel(corpus)
print tf_idf 
s = 0
for i in corpus:
    s += len(i)
print s 
```

```python
import os
cwd = os.getcwd()
sims = gensim.similarities.Similarity(cwd,tf_idf[corpus], num_features=len(dictionary))
print sims 
print(type(sims))
```

```python
score = 0
for index, article in enumerate(newsgroups_train.data[10001:11314]):
    query_doc = [w.lower() for w in word_tokenize(article)]
    query_doc_bow = dictionary.doc2bow(query_doc)
    query_doc_tf_idf = tf_idf[query_doc_bow]
    sims[query_doc_tf_idf]
    pred = newsgroups_train.target[:10000][np.argmax(sims[query_doc_tf_idf])]
    actual = newsgroups_train.target[10001+index]
    print pred, " - ", actual
    if pred == actual:
        score +=1
    
print "+=+=+=+=+=+=+=+=+=+=+=+=+=+=+="
print score, "out of" , len(newsgroups_train.target[10001:11314])
print "which is ", float(score)/float(len(newsgroups_train.target[10001:11314]))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
 
vectorizer = CountVectorizer(analyzer = "word",   \
                             tokenizer = word_tokenize,    \
                             preprocessor = review_to_words, \
                             stop_words = stopwords.words("english"),   \
                             max_features = 10000) 

train_data_features = vectorizer.fit_transform(newsgroups_train.data[:10000])

train_data_features = train_data_features.toarray()
```

```python
train_data_features.shape
```

```python
vocab = vectorizer.get_feature_names()
print vocab[:100]
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 200) 
forest = forest.fit(train_data_features, newsgroups_train.target[:10000])
```

```python
test_data_features = vectorizer.transform(newsgroups_train.data[10001:11314])
test_data_features = test_data_features.toarray()
result = forest.predict(test_data_features)
```
